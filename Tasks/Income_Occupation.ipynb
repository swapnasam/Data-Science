### Load the Income and Occupation text files

```r
income_df = read.table(file =  'Income.txt',  sep = '\t' , header = TRUE )
occupation_df = read.table(file =  'Occupations.txt',  sep = '\t' , header = TRUE )
income_df
occupation_df
```

## 1) Remove invalid occupations.

```r
# to get an idea of number of rows in the df
nrow(occupation_df)
```

```r
# to get an idea of different occupation title in the file and check whether the occupation title makes sense or not
distinct(occupation_df, Occupation)
```

```r
# To see how many rows have invalid Ocuupation
invalid_occupation <- subset(occupation_df, Occupation == "Married to Jim"| Occupation == "Materials Engineerr"| Occupation ==  "I look after my mother Sarah who is unwell")
invalid_occupation
```

#### Above code shows the distinct occupation in the text file. Occupation 'Married to Jim' and  'I look after my mother Sarah who is unwell' are invalid occupation. Therefore, we can remove the rows with these occupation.

#### However, the row with 'Materials Engineerr' is a typo and is not an invalid occupation. Therefore, we can correct it to Materials Engineer.


```r
# display the row with incorrectly spelled occupation
subset(occupation_df, Occupation =='Materials Engineerr')
```

```r
# Update the rows with Materials Engineerr to Materials Engineer
occupation_df <- within(occupation_df, Occupation[Occupation == 'Materials Engineerr'] <- 'Materials Engineer')

# check once again to see the update
subset(occupation_df, Occupation =='Materials Engineer')
```

```r
# Remove the invalid Occupation rows
occupation_df <- occupation_df[!(occupation_df$Occupation=="Married to Jim" | occupation_df$Occupation=="I look after my mother Sarah who is unwell"),]


# You can verify by again doing a check 
# subset(occupation_df, Occupation == "Married to Jim"| Occupation == "Materials Engineerr"| Occupation ==  "I look after my mother Sarah who is unwell")
```

## 2)Using occupation file, pull in income values, creating a file that shows "ID", "First_Name", "Last_Name", "Sex", "Occupation", "Income".

```r
# Perform a left outer join such that all the rowsin the Occupation file is considered and select only the required columns

merged_df = merge(x = occupation_df, y = income_df, by = "ID", all.x = TRUE) %>%
select(ID,"First_Name.x",Last_Name.x, "Sex.x","Occupation", "Income")

```

```r
merged_df
```

### We can notice some inconsistency in the data as there are repeating ID with different occupation. Therefore, these needs to be taken care or checked further. 

```r
# Rename the df as per the requirement 
names(merged_df)[names(merged_df) == "First_Name.x"] <- "First_Name"
names(merged_df)[names(merged_df) == "Last_Name.x"] <- "Last_Name"
names(merged_df)[names(merged_df) == "Sex.x"] <- "Sex"
```

```r
merged_df
```

## 3)Remove any NA incomes.


```r
# Remove all the rows where Income is NA
merged_df <- merged_df[!is.na(merged_df$Income),]
merged_df
```

## 4) Gather data so "Income", "Sex", and "Occupation" are values in a new column called "Variable"

```r
gathered_df <- merged_df %>%
gather(Variable, Value, Sex:Income)
gathered_df
```
