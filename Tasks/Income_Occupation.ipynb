#### Above code shows the distinct occupation in the text file. Occupation 'Married to Jim' and  'I look after my mother Sarah who is unwell' are invalid occupation. Therefore, we can remove the rows with these occupation.

#### However, the row with 'Materials Engineerr' is a typo and is not an invalid occupation. Therefore, we can correct it to Materials Engineer.


```r
# display the row with incorrectly spelled occupation
subset(occupation_df, Occupation =='Materials Engineerr')
```

```r
# Update the rows with Materials Engineerr to Materials Engineer
occupation_df <- within(occupation_df, Occupation[Occupation == 'Materials Engineerr'] <- 'Materials Engineer')

# check once again to see the update
subset(occupation_df, Occupation =='Materials Engineer')
```

```r
# Remove the invalid Occupation rows
occupation_df <- occupation_df[!(occupation_df$Occupation=="Married to Jim" | occupation_df$Occupation=="I look after my mother Sarah who is unwell"),]


# You can verify by again doing a check 
# subset(occupation_df, Occupation == "Married to Jim"| Occupation == "Materials Engineerr"| Occupation ==  "I look after my mother Sarah who is unwell")
```

## 2)Using occupation file, pull in income values, creating a file that shows "ID", "First_Name", "Last_Name", "Sex", "Occupation", "Income".

```r
# Perform a left outer join such that all the rowsin the Occupation file is considered and select only the required columns

merged_df = merge(x = occupation_df, y = income_df, by = "ID", all.x = TRUE) %>%
select(ID,"First_Name.x",Last_Name.x, "Sex.x","Occupation", "Income")

```

```r
merged_df
```

### We can notice some inconsistency in the data as there are repeating ID with different occupation. Therefore, these needs to be taken care or checked further. 

```r
# Rename the df as per the requirement 
names(merged_df)[names(merged_df) == "First_Name.x"] <- "First_Name"
names(merged_df)[names(merged_df) == "Last_Name.x"] <- "Last_Name"
names(merged_df)[names(merged_df) == "Sex.x"] <- "Sex"
```

```r
merged_df
```

## 3)Remove any NA incomes.


```r
# Remove all the rows where Income is NA
merged_df <- merged_df[!is.na(merged_df$Income),]
merged_df
```

## 5) Create a function that creates a column called fullname that combines first and last names.


```r
# The function'combine_name' takes firstname and lastname rows as input and returns the fullname
combine_name <- function (First_Name, Last_Name) {

paste(First_Name, Last_Name)
}
```

```r
final_df <- mutate(gathered_df, fullname =combine_name(First_Name,Last_Name) )
final_df
```

## 6) Plot overall distribution of income. Indicate the median on a line on the plot.


```r
Income_df_plot <- final_df %>%

                    filter(Variable == 'Income')

Income_df_plot
```

#### the Value field is of type character, therefore , we need to change its type to plot the distribution

```r
Income_df_plot$Value <- as.numeric(Income_df_plot$Value)
glimpse(Income_df_plot)
```

```r
# Draw histogram
ggp <- ggplot(data.frame(Income_df_plot), aes(Income_df_plot$Value)) +    
       geom_histogram(bins = 8,  # Draw histogram
                      col="red", 
                      aes(fill=..count..),
                      alpha = .2)  +
      ggtitle("Distribution of Income") +
      theme(plot.title = element_text(hjust = 0.5))  +
      labs(x="Income", y="Count") + 

      ylim(c(0,8))+  scale_fill_gradient("Count", low="green", high="red")


# Add line for median
ggp +                                     
  geom_vline(xintercept = median(Income_df_plot$Value),      
             col = "blue",
             lwd = 3) 

# Save the image
ggsave("Income_distribution_plot.png")
```

```r
overall_median = median(Income_df_plot$Value)
overall_median
```

## 7) Plot median income by occupation and compare with the overall median income.

```r
#lets  take the merged_df from step 3 

# Lets group based on name 
grouped_df <-merged_df %>% group_by(Occupation)
group_df <-grouped_df %>% summarise(Income = median(Income))

group_df
```

```r
group_df <-group_df %>%
        mutate(Overall_Median = overall_median)
group_df
```

```r
## To plot bar plot for Overall median and Occupation median

Block_name <-c("Occupation Median","Overall Median")
colors <- c("blue", "red")



count <-c(group_df$Income,group_df$Overall_Median)
Values <- matrix(count, nrow = 2, ncol = 7, byrow = TRUE) # to make as a matrix
#par(mar = c(9, 4, 4, 2) + 1)

png(file="Income_Occupation_bar_plot.png")
par(mar=c(15,8,8,8))#add room for  labels

barplot(Values ,names.arg=group_df$Occupation,col= colors,ylab = "Income", main ="Overall Median Income V/S Occupational Median Income",las = 2, beside =TRUE) ###las for space for label
legend("topright", Block_name, cex = .9, fill = colors)
dev.off()

# To display in jupyter lab
par(mar=c(15,8,8,8))
barplot(Values ,names.arg=group_df$Occupation,col= colors,ylab = "Income", main ="Overall Median Income V/S Occupational Median Income",las = 2, beside =TRUE) ###las for space for label
```
