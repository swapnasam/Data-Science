```python
#Import Libraries
import os
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
import cv2
import gc
```

```python
# Resizing and Converting to RGB
def load_and_preprocess_image(path):
    try:
        image = cv2.imread(path)
        image = cv2.resize(image, (224,224))
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        return image
    except Exception as e:
        print(str(e))
```

```python
#Read Stanford Car TRAIN paths folders
TRAIN_BASEPATH = "../input/stanford-car-dataset-by-classes-folder/car_data/car_data/train/"
train_paths = []
    
for d in os.listdir(TRAIN_BASEPATH):
    train_paths.append((TRAIN_BASEPATH+d, d))	
print(train_paths[0])
```

```python
#Read Stanford Car TEST paths folders
TEST_BASEPATH = "../input/stanford-car-dataset-by-classes-folder/car_data/car_data/test/"

test_paths = []
    
for d in os.listdir(TEST_BASEPATH):
    test_paths.append((TEST_BASEPATH+d, d))	
print(test_paths[0])
```

```python
#Garbage Collect
for i in range(1,50):
    gc.collect()
```

```python
#Stanford Car TRAIN Dataset
X_train = []
y_train = []
#Iterate each folder to obtain image and label
for path, label in train_paths:
    #print(path,label)
    for image_path in os.listdir(path):
        #print(image_path)
        image = load_and_preprocess_image(path+"/"+image_path)
        
        X_train.append(image)
        y_train.append(label)
```

```python
#Garbage Collect
for i in range(1,50):
    gc.collect()
```

```python
#Stanford Car TEST Dataset
X_test = []
y_test = []
#Iterate each folder to obtain image and label
for path, label in test_paths:
    #print(path,label)
    for image_path in os.listdir(path):
        #print(image_path)
        image = load_and_preprocess_image(path+"/"+image_path)
        
        X_test.append(image)
        y_test.append(label)
```

```python
#Garbage Collect
for i in range(1,50):
    gc.collect()
```

```python
#Convert to Numpy Array
X_train = np.array(X_train)
X_test = np.array(X_test)
```

```python
#Convert to Numpy Array
y_train=np.array(y_train)
y_test=np.array(y_test)
```

```python
#Using ResNet101v2, removing the fully connected layer and extracting the features
from keras import applications

input_shape = (224,224,3)
extraction_model = applications.resnet_v2.ResNet101V2(include_top=False, weights='imagenet', input_shape=input_shape, pooling='avg')
  
train_features = extraction_model.predict(X_train)
test_features = extraction_model.predict(X_test)
```

```python
#Perform Logistic Regression on the Feature Extracted
#Using L-BFGS solver with L2 Penalty and Maximum Iterations of 2000
clf2 = LogisticRegression(penalty='l2',solver='lbfgs',max_iter=2000)
clf2.fit(train_features, np.squeeze(y_train))
acc2 = clf2.score(test_features, np.squeeze(y_test))
print( 'The classifier got an accuracy of : {}'.format(acc2))
```
